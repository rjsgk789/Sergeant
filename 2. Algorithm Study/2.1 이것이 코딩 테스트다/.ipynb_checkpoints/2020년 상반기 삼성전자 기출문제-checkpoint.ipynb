```python
# p.402 아기 상어
'''
반복
    자기보다 작은 물고기가 몇마리인지 count

    0마리이면 끝
    1마리이면 바로 먹으러 가
    2마리 이상이면 가장 가까운 물고기 먹으러 가 (여러마리면, 가장 위, 가장 왼쪽에 있는 물고기를 먹으러 가)
    -->> 먹으러가는 경로, 최단거리 구하기는 bfs로 구현
    
    아기상어가 먹은 물고기 수 count하고 자기 크기만큼 먹으면 크기+1, count초기화
    먹힌 물고기 칸은 지나간 후에 0으로 바뀌어
'''

from collections import deque


# bfs 함수 구현 (전체 맵 map, 아기 상어의 좌표 x, y, 아기 상어의 크기 size 입력받기) (움직인 횟수, 도착 인덱스 반환 - (r,c)형태)
def bfs(MAP, r, c, baby_size):
    # 변수 설정
    dr = [-1,0,0,1]
    dc = [0,-1,1,0]
    n = len(MAP)
    distance = [[-1] * n for _ in range(n)]
    distance[r][c] = 0
    
    # deque() 객체 생성
    queue = deque()
    # 현재 좌표 넣기
    queue.append((r, c))
    # deque 반복
    while queue:
        r, c = queue.popleft()
        # 위 왼 오 아래 순으로 반복
        for i in range(4):
            nr = r+dr[i]
            nc = c+dc[i]
            # 범위 벗어나면 continue
            if nr<0 or nc<0 or nr>=n or nc>=n:
                continue
            
            now = MAP[nr][nc]
            # 물고기 크기가 아기상어보다 크면 continue
            if now != 9 and now > baby_size:
                continue
            # 위에 최단거리 이중리스트 만들고 첫 방문시에만(초깃값 설정) 최단거리 = 이전거리+1
            if distance[nr][nc] == -1:
                distance[nr][nc] = distance[r][c] + 1
                queue.append((nr, nc))
            
            # 물고기 크기가 아기상어보다 작다 -> 바로 return 해버리기
            if now != 0 and now < baby_size:
                return distance[nr][nc], (nr, nc)
    

n = 4
MAP = [
    [4,3,2,1],
    [0,0,0,0],
    [0,0,9,0],
    [1,2,3,4]
]

baby_size = 2
eat_count = 0


total_move = 0

# while True
while(True):
    can_eat_count = 0
    can_eat_list = []
    
    # row 반복
    for r in range(n):
        # column 반복
        for c in range(n):
            now = MAP[r][c]
            if now==0:
                continue
            elif now==9:
                baby_rc = (r, c)
                baby_before_index = (r, c)
                continue
            # 아기상어의 크기보다 작으면
            if MAP[r][c] < baby_size:
                # count+=1, 먹을 수 있는 물고기 리스트(can_eat)에 index(r, c) 추가
                can_eat_count += 1
    
    # if count==0
    if can_eat_count == 0:
        # 끝
        break
    # if count==1
    else:
        # dfs로 거리, 위치 찾기
        move, place = bfs(MAP, baby_rc[0], baby_rc[1], baby_size)
        MAP[baby_before_index[0]][baby_before_index[1]] = 0
        MAP[place[0]][place[1]] = 9
    # if count>=2
        # dfs로 가장 가까운 물고기까지 거리, 위치 찾기
    
    # 거리만큼 총 이동거리 +, 아기상어 위치이동, eat_count+=1
    total_move += move
    eat_count += 1
    # 만약 size == eat_count -> size += 1, eat_count = 0
    if baby_size == eat_count:
        baby_size += 1
        eat_count = 0


# 총 이동거리 출력
print(total_move)
```

```python
# Chapter1.2 - chipotle의 주문 데이터 분석하기
# Step1. 데이터의 기초 정보 살펴보기
import pandas as pd

file_path = 'data/chipotle.tsv'

chipo = pd.read_csv(file_path, sep = '\t')

print(chipo.shape)
print("------------------------------")
print(chipo.info())
```

```python
chipo.head(10)
```

```python
print(chipo.columns)
print("------------------------------")
print(chipo.index)
```

```python
'''
order_id는 의미가 없은 수치형 데이터이므로 문자열로 바꾼다.
csv파일.describe() -> 수치형 데이터들의 기초 통계량 표시

item_price는 수치형 데이터로 이용하려면 전처리가 필요함. (-> Step3)
'''
chipo['order_id'] = chipo['order_id'].astype(str)
chipo.describe()
```

```python
'''
order_id와 item_name의 개수
범주형 피쳐 -> unique() 함수 사용해서 범주 개수 파악
'''
print(len(chipo['order_id'].unique()))
print(len(chipo['item_name'].unique()))
```

```python
# Step2. 탐색과 시각화
'''
가장 많이 주문한 아이템 top10
DataFrame['column'] -> series라는 객체 반환
value_counts() -> series 객체에 적용
'''

item_count = chipo['item_name'].value_counts()[:10]
print(chipo['item_name'].value_counts())
for idx, (val, cnt) in enumerate(item_count.iteritems(), 1):
    print("Top", idx, ":", val, cnt)
```

```python
'''
아이템별 주문 개수와 총량

groupby() 함수 이용 -> 데이터프레임에서 특정 피처를 기준으로 그룹을 생성
'''

order_count = chipo.groupby('item_name')['order_id'].count()
order_count
```

```python
item_quantity = chipo.groupby('item_name')['quantity'].sum()
item_quantity
```

```python
'''
시각화

아이템별 주문의 총량 -> 막대그래프로 시각화
tolist()와 np.arrange()를 이용. 0~50으로 50개의 아이템 표현 (x)

'''

import numpy as np
import matplotlib.pyplot as plt

item_name_list = item_quantity.index.tolist()
x_pos = np.arange(len(item_name_list))
order_cnt = item_quantity.values.tolist()

plt.bar(x_pos, order_cnt, align='center')
plt.ylabel('ordered_item_count')
plt.title('Distribution of all ordered item')

plt.show()
```

```python
# Step3. 데이터 전처리
'''
수치적 특징을 탐색했음. - item_price 피처의 요약 통계를 구할 수 없었다. (문자열 타입이라서)
이 것을 전처리 해보자
'''

print(chipo.info())
print('-------------------------')
chipo.head()
```

```python
'''
$기호 없애기
'''

chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
chipo.describe()
```

```python
'''
item_price의 타입이 float64로 바뀜
'''
chipo.info()
```

```python
# Step 4. 탐색적 분석
'''
주문당 평균 계산금액 출력하기
order_id로 그룹 생성 -> item_price 피처에 sum() 함수 적용 -> mean() 함수 추가
주문서 별로 결제금액을 계산한 후, 평균을 나타낸 거지
'''
chipo.groupby('order_id')['item_price'].sum().mean()
```

```python
'''
한 주문에 10달러 이상 지불한 주문 번호(order_id) 찾기
'''

chipo_orderid_group = chipo.groupby('order_id').sum()
result = chipo_orderid_group[chipo_orderid_group.item_price >= 10]

print(result)
print(result.index.values)
```

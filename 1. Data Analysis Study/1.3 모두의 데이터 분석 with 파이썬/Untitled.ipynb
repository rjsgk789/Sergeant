```python
import matplotlib.pyplot as plt
plt.title('plt')
plt.plot([1, 2, 3, 4], [12, 43, 25, 15], color='red', ls=':', label='cool')
plt.plot([10, 20, 30, 40, 50], color='b', ls='--', label='asc')

plt.legend(loc=1) # 범례 표시 (cool, asc)
plt.show()
```
